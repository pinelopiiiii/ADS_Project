Load packages

```python
import random
from random import randint
import pandas as pd
from faker import Faker
```

Faker on Dutch

```python
fake = Faker("nl_NL")
```

```python
#Create data with fake and random
data = {
    'ID': [fake.random_int(min=1000, max=9999) for _ in range(10)],
    'TXT1_vacation': [fake.text() for _ in range(10)],
    'TXT2_summer': [fake.text() for _ in range(10)],
    'TXT3_apology': [fake.text() for _ in range(10)],
    'TXT4_complain': [fake.text() for _ in range(10)],
    'csp_un_bl1_I_avoid': [random.randint(0, 100) for _ in range(10)],
    'csp_av_bl1_I_avoid': [random.randint(0, 100) for _ in range(10)],
    'csp_un_bl2_I_avoid': [random.randint(0, 100) for _ in range(10)],
    'csp_av_bl2_I_avoid': [random.randint(0, 100) for _ in range(10)],
    'csm_un_bl1_I_avoid': [random.randint(0, 100) for _ in range(10)],
    'csm_un_bl2_I_avoid': [random.randint(0, 100) for _ in range(10)]
}

#Create df
df = pd.DataFrame(data)
df
```

```python
#Save df as csv file
df.to_csv('fake_data.csv', index=False)
```
